```python
import matplotlib.pyplot as mpl
import math as m
import numpy as np
import random as rnd

### демонстрация работы численных методов решения СЛУ

def Method_Jacobi(m, f, dim):
    x_0 = [0] * dim
    x_1 = []

    for n in range(1000):
        for j in range(dim):
            x_1.append(f[j]/m[j][j])
            for i in range(dim):
                if i != j:
                    x_1[j] -= m[j][i]/m[j][j]*x_0[i]
        x_0 = x_1
        x_1 = []

    return x_0


def Method_Nekrasov(m, f, dim):
    x_0 = [0] * dim

    for n in range(1000):
      for j in range(dim):
            x_0[j] = f[j]/m[j][j]
            for i in range(dim):
                if i != j:
                   x_0[j] -= m[j][i]/m[j][j]*x_0[i]

    return x_0


A = [[2, -1, 0], [-1, 2, -1], [0, -1, 2]]
F = [1, 2, 3]
dim1 = 3
# answer 2.5 4 3.5

print('метод Якоби', Method_Jacobi(A, F, dim1))
print("Для матрицы в системе линейных уравнений: ")
for i, a in enumerate(A):
  print(a, F[i])

print('метод Зейделя-Некрасова', Method_Nekrasov(A, F, dim1))
print("Для матрицы в системе линейных уравнений: ")
for i, a in enumerate(A):
  print(a, F[i])



###

#### демонстрация работы численных методов решения систем линейных неоднородных уравнений

def matrix(x):
    return(x[0] + 0.5 * x[1]**2 - 4, x[1] + 0.5 * x[2]**2 - 4, 0.5 * x[0]**2 + x[2] - 4)

def jacobian(x):
    xyz = x[0] * x[1] * x[2]
    return([1 - xyz/(1 + xyz), xyz * x[1] * x[2]/(1+xyz) - x[1], x[1] * x[2]/(1 + xyz)],
           [x[0] * x[2]/(1 + xyz), 1 - xyz/(1 + xyz), -x[2]/(1 + xyz)],
            [-x[0]/(1 + xyz), xyz/(1 + xyz), 1 / (1 + xyz)])

def NonLinear1(x_0, i):

    for i in range(100):
        jac = jacobian(x_0)

        f = matrix(x_0)

        delta = [0, 0, 0]
        for j in range(3):
            for k in range(3):
                delta[j] += jac[j][k]*f[k]
        for j in range(3):  x_0[j] -= delta[j]

    return x_0

x_0 = [2, 2, 2]
x = NonLinear1(x_0, 1)

print("\nСистема нелинейных уравнений:")
print(" x_0 + 0.5 * x_1 ^ 2 - 4 = 0\n", "x_1 + 0.5 * x_2 ^ 2 - 4 = 0\n", "x_2 + 0.5 * x_0 ^ 2 - 4 = 0")
print('Решение системы нелинейных уравнений', x)


####

# метод прогонки, или метод тридиагональной матрицы. Для решения СЛУ

def Progon(mat, f):

    x = [0 for i in range(5)]
    v = [0 for i in range(5)]
    u = [0 for i in range(5)]

    v[0] = -mat[0][1] / mat[0][0]
    u[0] = f[0]/mat[0][0]

    for i in range(1, 4):
        v[i] =  mat[i][i+1] / ( - mat[i][i] - mat[i][i-1]*v[i-1])
        u[i] = (mat[i][i-1]*u[i-1] - f[i-1]) / (-mat[i][i] - mat[i][i-1]*v[i-1])

    i = 4
    v[i] = 0
    u[i] = (mat[i][i-1]*u[i-1] - f[i-1]) / (-mat[i][i] - mat[i][i-1]*v[i-1])

    x[4] = u[4]

    for i in range(4, 0, -1):
        x[i-1] = v[i-1]*x[i] + u[i-1]

    return x

A = ([117, 2, 0, 0, 0],
     [2, 117, 2, 0, 0],
     [0, 2, 117, 2, 0],
     [0, 0, 2, 117, 2],
     [0, 0, 0, 2, 117])

F = (119, 121, 121, 121, 119)
#1 1 1 1 1

print('\nМетод прогонки', Progon(A, F))
print("Для тридиагональной матрицы в системе линейных уравнений: ")
for i, a in enumerate(A):
  print(a, F[i])



```

Функция Jacobi решает систему линейных уравнений с помощью метода Якоби.

Входные параметры:

mat - матрица A системы уравнений размерности k×k.
f - вектор правой части системы уравнений размерности k.
k - размерность системы уравнений.
Решение системы уравнений методом Якоби происходит следующим образом:

Инициализируем начальное приближение x_0 вектором нулей размерности k.
Повторяем следующее для заданного числа итераций (в данном случае 1000):
Для каждого уравнения j от 0 до k-1:
Вычисляем новую компоненту x_1[j] путем деления соответствующего элемента вектора правой части f[j] на соответствующий диагональный элемент матрицы mat[j][j].
Для каждой компоненты i от 0 до k-1 (исключая j):
Из x_1[j] вычитаем вклад компоненты i (коэффициент mat[j][i]/mat[j][j]) умноженный на предыдущую компоненту x_0[i].
Обновляем значение x_0 на x_1.
Очищаем x_1 для следующей итерации.
Возвращаем приближенное решение системы уравнений, хранящееся в x_0.

В методе Якоби значения x_0 используются только после завершения итерации для вычисления новых значений компонент вектора x (x_1). То есть, все компоненты вектора x_1 вычисляются с использованием предыдущих значений x_0. После завершения итерации, значения x_1 становятся новым значением x_0, и процесс повторяется.

В методе Зейделя-Некрасова, значения x_0 обновляются непосредственно внутри итерационного цикла, при вычислении каждой компоненты x_0[j]. Это означает, что при вычислении очередной компоненты вектора x_0, уже обновленные значения других компонент могут быть использованы в расчетах. Таким образом, значения компонент вектора x могут обновляться по мере выполнения итераций, а не только после каждой итерации.
