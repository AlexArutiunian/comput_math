```python
import numpy as np
import matplotlib.pyplot as plt


def center_point(f, x, h):

    # for 1st order of approx
    return (f(x + h) - f(x)) / (h)

def front_point(f, x_start, h):
    return (-3 * f(x_start) + 4 * f(x_start + h) - f(x_start + 2 * h)) / (2 * h)

def back_point(f, x_end, h):
     return ( (1.5) * f(x_end) - 2 * f(x_end - h) + 0.5 * f((x_end - 2 * h))) / (h)

def y(x):
    return x ** 2

def rhs(x):
    return 2 * x

x_start = 0
x_end = 1
num_points = 1000
# Различные значения h в логарифмической шкале
h_values = np.logspace(-6, -1, num_points)

errors = []

for h in h_values:
    x_ = np.linspace(x_start, x_end, int((x_end - x_start) / h) + 1)
    x_values = np.linspace(x_start + h, x_end - h, len(x_) - 2)

    y_der_by_approx = [front_point(y, x_start, h)]

    for x in x_values:
        y_der_by_approx.append(center_point(y, x, h))

    y_der_by_approx.append(back_point(y, x_end, h))

    y_der_teor = [rhs(x) for x in x_]

    error = np.max(np.abs(np.subtract(y_der_by_approx, y_der_teor)))
    errors.append(error)

plt.plot(h_values, errors, label='log(Errors(h)) for 1st der 1st order')
plt.xlabel('h')
plt.ylabel('Error')
plt.yscale('log')
plt.xscale('log')
plt.legend()
fit = np.polyfit(np.log10(h_values), np.log10(errors), 1)
slope = fit[0]
plt.show()
```
